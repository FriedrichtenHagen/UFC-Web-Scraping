```python
import sqlite3
import csv
```

```python
with open ('ufc_events_sample_2.csv','r') as ufc_events_csv:
    reader = csv.DictReader(ufc_events_csv)
    columns = []
    rows = []
    for row in reader:
        rows.append((row['event_id'],row['event_name'],row['event_date'],row['event_city'],row['event_state'],row['event_country']))
```

```python
conn = sqlite3.connect('ufc_database.db')
c = conn.cursor()
```

```python
c.execute("""DROP TABLE ufc_events""")
```

```python
c.execute("""CREATE TABLE ufc_events(
event_id integer,
event_name text,
event_date date,
event_city text,
event_state text,
event_country text)""")

conn.commit()
```

```python
for row in rows:
    c.execute("""INSERT INTO ufc_events
    VALUES(?,?,?,?,?,?);""", row)
```

```python
c.execute("""SELECT *
FROM ufc_events""")
```
