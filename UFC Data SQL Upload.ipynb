```python
import sqlite3
import csv
import os
```

```python
event_rows = []
fight_rows = []
```

```python
with open ('ufc_event_data.csv','r') as ufc_events_csv:
    reader = csv.DictReader(ufc_events_csv)
    for row in reader:
        event_rows.append((row['event_name'],row['event_date'],row['event_city'],row['event_state'],row['event_country']))
```

```python
conn = sqlite3.connect('ufc_database.db')
c = conn.cursor()
```

```python
c.execute("""DROP TABLE IF EXISTS ufc_events""")
c.execute("""DROP TABLE IF EXISTS ufc_fights""")
```

```python
c.execute("""CREATE TABLE ufc_events(event_id INTEGER PRIMARY KEY AUTOINCREMENT,
event_name TEXT,
event_date DATE,
event_city TEXT,
event_state TEXT,
event_country TEXT)""")
```

```python
c.execute("""CREATE TABLE ufc_fights(fight_id INTEGER PRIMARY KEY AUTOINCREMENT,
event_name TEXT,
referee TEXT, 
f_1 TEXT, 
f_2 TEXT, 
winner TEXT, 
num_rounds INTEGER, 
weight_class TEXT, 
result TEXT, 
result_details TEXT, 
finish_round INTEGER, 
finish_time TEXT, 
f_1_kd INTEGER,
f_2_kd INTEGER,
f_1_sig_strikes_att INTEGER,
f_1_sig_strikes_succ INTEGER,
f_2_sig_strikes_att INTEGER,
f_2_sig_strikes_succ INTEGER,
f_1_strikes_att INTEGER,
f_1_strikes_succ INTEGER,
f_2_strikes_att INTEGER,
f_2_strikes_succ INTEGER,
f_1_td_att INTEGER,
f_1_td_succ INTEGER,
f_2_td_att INTEGER,
f_2_td_succ INTEGER,
f_1_sub_att INTEGER,
f_2_sub_att INTEGER,
f_1_rev INTEGER,
f_2_rev INTEGER,
f_1_ctrl_time TEXT,
f_2_ctrl_time TEXT,
fight_url TEXT)""")
```

```python
for row in fight_rows[::-1]:
    c.execute("""INSERT INTO ufc_events (event_name,
 referee,
 f_1,
 f_2,
 winner,
 num_rounds,
 weight_class,
 result,
 result_details,
 finish_round,
 finish_time,
 f_1_kd,
 f_2_kd,
 f_1_sig_strikes_att,
 f_1_sig_strikes_succ,
 f_2_sig_strikes_att,
 f_2_sig_strikes_succ,
 f_1_strikes_att,
 f_1_strikes_succ,
 f_2_strikes_att,
 f_2_strikes_succ,
 f_1_td_att,
 f_1_td_succ,
 f_2_td_att,
 f_2_td_succ,
 f_1_sub_att,
 f_2_sub_att,
 f_1_rev,
 f_2_rev,
 f_1_ctrl_time,
 f_2_ctrl_time,
 fight_url)
    VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?);""", row)
```
